```python
driver = webdriver.Chrome()
driver.get('https://www.wadiz.kr/web/wreward/category/292?keyword=&endYn=Y&order=support')
```

```python
num_scroll = 15
while num_scroll:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight-1000);")
    time.sleep(2)
    num_scroll -= 1
time.sleep(1)
```

```python

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

browser = webdriver.Chrome('./chromedriver.exe')

data_list = soup.select('div.ProjectCardList_list__1YBa2 > div > div > div > div')
game =[]
for data in data_list:
    title = data.select('strong')[0].text.strip()
    print(title)
    game.append([title])
```

```python
pd_data = pd.DataFrame(game, columns=['wadiz game_hobby'])
pd_data
```
